```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame with random data representing 2-way transmission delay for different distances
df = pd.read_csv("bluetooth_delay.csv")
df = df * 1000
# Create box and whisker plot
plt.figure(figsize=(10, 6))
df.boxplot()
plt.title('2-way BLE Transmission Delay for Battlebot', fontsize=14)
plt.ylabel('Transmission Delay (ms)', fontsize=14)
plt.xlabel('Distance (feet)', fontsize=14)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV
df = pd.read_csv('bluetooth_connection.csv')

# Assuming the row of connection times is labeled 'Connection Times'
connection_times = df.loc[0]

# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(connection_times, bins=20, edgecolor='black')
plt.title('Distribution of Connection Times', fontsize=14)
plt.xlabel('Time (s)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([5, 7.5, 10, 17.5, 22.5])
y = np.array([3340, 6760, 9410, 11650, 13450])

# Transform data into linear speed (mph)
r = 50 # mm
c = np.pi * 2 * r # mm
y = y * c # mm/min
y = y * 60 # mm/hr
y = y / 1.609e6

# Plot the original data points and the fitted curve
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o', linestyle='-', label='Tachometer Data')
plt.axhline(y=150, color='r', linestyle='--', label='150mph (Target Speed)')
plt.title('Weapon Tip Speed vs. Throttle Percentage')
plt.xlabel('Throttle (%)')
plt.ylabel('Weapon Tip Speed (mph)')
plt.legend()
plt.grid(True)
plt.show()
```
